```python
import pandas as pd
```

```python
class Dial():
    def __init__(self, list_length, position=0):
        self.list_length = list_length
        self.linked_list = range(0, list_length)
        self.position = position
        
    def click(self, click):
        click_direction, click_distance = click[0], int(click[1:])
        
        if click_direction == 'L':
            click_distance = click_distance * -1

        start = self.position
        self.position = self.position + click_distance
        self.position = self.state()
        
        zero_crossings = self.check_zero_crossings(start, click_distance)   

        return self.state(), zero_crossings
    
    def state(self):
        return self.linked_list[self.position%self.list_length]
    
    def set_state(self, state):
        self.position = state
        
    def check_zero_crossings(self, start, change):
        new = start + change
        extra = 0
        
        # if new < 0 need to tweak a little
        if new < 0:
            # stops at zero, add 1
            if new%self.list_length == 0:
                extra = 1
            # did it start at 0?
            if start == 0:
                # moved less than a dial so its zero
                if abs(change) < self.list_length:
                    return 0
                else:
                    # change >= 100
                    # eg change = -199 -> new = -199, so it should be 1
                    # eg change = -279 -> new = -279 so it should be 2
                    # eg change = -100 -> new = -100 so it should be 1 and extra = 1
                    return abs(new)//self.list_length + extra
            else:
                return abs(new//self.list_length) + extra
        
        else:
            # stops at zero, add 1
            if new == 0:
                extra = 1
            # if its > length, then this gives how many times it crosses 0
            # if new == 0 this should give zero
            crosses = new//self.list_length
            return crosses + extra
            
```

```python
my_dial = Dial(100, 50)
```
